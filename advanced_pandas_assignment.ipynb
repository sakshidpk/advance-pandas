7. Pivot Table Creation
Create a pivot table showing:
Rows → Region
Columns → Product
Values → Total Sales
Which region sells the most laptops?

```python
pivot_table_sales = pd.pivot_table(df, values='Sales', index='Region', columns='Product', aggfunc='sum')
display(pivot_table_sales)
```

```python
region_most_laptop_sales = pivot_table_sales['Laptop'].idxmax()
max_laptop_sales = pivot_table_sales['Laptop'].max()

print(f"The region that sells the most laptops is: {region_most_laptop_sales} with total sales of {max_laptop_sales}")
```

9. Discount Impact Study
Analyze the relationship between discount and profit.
Tasks:
Calculate the average profit for each discount level
Identify whether higher discounts lead to lower profit

```python
avg_profit_by_discount = df.groupby('Discount')['Profit'].mean().reset_index()
display(avg_profit_by_discount.sort_values(by='Discount'))
```

### Discount Impact Study Insights

By analyzing the average profit for each discount level, we can observe the following relationship:

*   **Inverse Relationship:** Generally, as the discount percentage increases, the average profit tends to decrease. This is evident from the sorted data, where higher discounts (e.g., 0.15, 0.20) often correspond to lower average profits, and sometimes even negative profits.

*   **Zero Discount Profitability:** Products sold without any discount (0.00 discount) typically yield the highest average profits, indicating that discounts, while potentially driving sales volume, directly erode profit margins.

*   **Risk of High Discounts:** High discount levels (e.g., 0.20) can lead to significantly reduced profitability, and in some cases, result in average losses, as seen in the data. This suggests that while discounts can attract customers, aggressive discounting needs to be carefully managed to avoid negatively impacting the bottom line.

10. End-to-End Data Wrangling Task

Perform the following workflow:

Filter the dataset where Sales > 3000

Group data by Region and Product

Calculate total sales

Reshape the result into a pivot table

Explain the business insights obtained from this analysis.

### 10.1 Filter the dataset where Sales > 3000

```python
df_filtered = df[df['Sales'] > 3000]
display(df_filtered.head())
```

### 10.2 Group data by Region and Product and Calculate total sales

```python
grouped_sales = df_filtered.groupby(['Region', 'Product'])['Sales'].sum().reset_index()
display(grouped_sales.head())
```

### 10.3 Reshape the result into a pivot table

```python
pivot_table_filtered_sales = pd.pivot_table(grouped_sales, values='Sales', index='Region', columns='Product', aggfunc='sum', fill_value=0)
display(pivot_table_filtered_sales)
```

### 10.4 Business Insights from this Analysis

This analysis, focusing on high-value sales (Sales > 3000), grouped by region and product, and presented in a pivot table, provides several key business insights:

*   **High-Value Product Dominance:** Electronics products (Laptops, Phones, Monitors, Tablets) consistently drive high sales across all regions when individual sales exceed $3000. This indicates that these are premium products or are frequently bought in quantities that result in high transaction values.

*   **Regional Strengths in Premium Products:** Each region exhibits its own strengths in high-value products:
    *   **East:** Shows strong performance across various Electronics, especially in Laptops and Phones, and also for Accessories like Keyboards and Headphones, suggesting a robust market for both high-end electronics and bulk purchases of accessories.
    *   **North:** While Electronics are present, there's also a significant contribution from Office products, particularly Printers, at high sales values. This might indicate a strong B2B or institutional market for office equipment.
    *   **South:** Displays a balanced contribution from various Electronics and Accessories, with a notable presence of Mouse and Headphones in high-value transactions, possibly indicating a consumer market for gaming or high-performance peripherals.
    *   **West:** Also shows strong Electronics sales. Keyboards and Phones stand out in the high-value segment, potentially pointing to specific customer preferences or market demand for these products.

*   **Targeted Marketing Opportunities:** By understanding which high-value products are most successful in each region, businesses can tailor their marketing campaigns, inventory management, and promotional strategies. For example, the East region could be targeted with promotions for premium laptops and phones, while the North could see campaigns for advanced office equipment.

*   **Inventory and Supply Chain Optimization:** Insights into regional high-value product demand can help optimize inventory levels and supply chain logistics, ensuring that popular high-ticket items are readily available in regions where they perform best.

*   **Identifying Niche Markets:** The presence of certain products (e.g., Office products in the North) in the high-value segment can help identify niche markets that might require specialized attention or product offerings.
